## Nested Lists (2D Lists) — Quick Scaffolding

A **nested list** (also called a **2D list**) is a list where each item is itself a list.
We often use nested lists to represent **grids** like seating charts, game boards, or tables.

### 1) A small example (2 rows × 3 columns)

```python
grid = [
    ["EMPTY", "EMPTY", "EMPTY"],
    ["EMPTY", "EMPTY", "EMPTY"]
]
```
### 2) Building a grid with loops (recommended)

To build a grid, we usually make rows one at a time and add them to the chart.
```
def make_chart(rows, cols):
    chart = []
    for r in range(rows):              # outer loop = rows
        row = []
        for c in range(cols):          # inner loop = columns
            row.append("EMPTY")
        chart.append(row)
    return chart
```

### 3) Printing a grid neatly (nested loops)

```
def print_chart(chart):
    for r in range(len(chart)):
        for c in range(len(chart[0])):
            print(chart[r][c], end=" ")
        print()  # new line after each row
```

### 4) Updating one cell
```
chart[2][3] = "Ava"   # row 2, col 3
```

### 5) Common reminders

- The outer loop usually goes through the rows

- The inner loop usually goes through the columns

- Use len(chart) for number of rows

- Use len(chart[0]) for number of columns (assuming all rows are same length)

### F1. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# F1 code here

def assign_seats(chart,row,col,name):
  chart[row][col]=name

def print_chart(chart):
  for row in range(len(chart)):
        for col in range(len(chart[0])):
            print(chart[row][col], end=" ")
        print()

grid = []
for row in range(4):              # outer loop = rows
    row = []
    for col in range(5):          # inner loop = columns
        row.append("EMPTY")
    grid.append(row)

assign_seats(grid, 0,1, "Alison")
assign_seats(grid, 2,2, "Leah")
assign_seats(grid, 3,3, "Ruchama")
assign_seats(grid, 0,4, "Reizi")
assign_seats(grid, 1,1, "Chana")
assign_seats(grid, 3,1, "Shaked")
print_chart(grid)

values=0
for row in range(len(grid)):
  for col in range(len(grid[0])):
    if grid[row][col]=="EMPTY":
      values+=1

print("There are ", values, " unassigned seats left.")
````

### F2. Temperature Grid (nested list) (2D list + loops + functions)
Create a weekly temperature tracker as a nested list.

**Requirements:**
- Build a 7×3 grid (7 days, 3 readings per day: morning/afternoon/evening) filled with `0`
- Write a function `set_temp(grid, day, time_index, temp)` that updates one value

- Write a function `print_grid(grid)` that prints each day neatly
- Fill in at least **10** temperatures (your choice)
- Use nested loops to compute:
  - the **overall average temperature**
  - the **highest single reading** in the grid

**Optional extension:** Count how many readings are above a threshold (e.g., above 75).


````python
# F2 answer here
def set_temp(grid, day, time_index, temp):
 grid[day-1][time_index]=temp #minus one because day 7 for index is 6 and so on

def print_grid(grid):
  for day in range(len(grid)):
    print(f"Day {day+1}: ", end="") #displays it nicely using an f string to tell the user what day it is
    for time_index in range(len(grid[0])):
      print(grid[day][time_index], end=" ")
    print()




grid = []
for row in range(7):              # outer loop = rows
    row = []
    for col in range(3):          # inner loop = columns
        row.append(0)
    grid.append(row)


set_temp(grid, 1, 0, 65)
set_temp(grid, 2, 1, 75)
set_temp(grid, 3, 2, 50)
set_temp(grid, 4, 0, 60)
set_temp(grid, 5, 1, 87)
set_temp(grid, 6, 2, 45)
set_temp(grid, 7, 0, 30)
set_temp(grid, 2, 1, 90)
set_temp(grid, 4, 2, 63)
set_temp(grid, 6, 1, 98)
print_grid(grid)

total_temp=0
total_readings=0
highest_reading=grid[0][0]

for day in range(len(grid)):#loops through each temp
    for time_index in range(len(grid[0])):
      value= grid[day][time_index]
      if value!=0:#Makes sure a reading was done of the temp
        total_temp+=grid[day][time_index]
        total_readings+=1
        if value>=highest_reading: #calcukates the highest reading
          highest_reading=value

average_temp=total_temp/total_readings
print(average_temp)
print(highest_reading)
````

### F3. Mini Gradebook (nested list) (2D list + loops + functions)
Create a mini gradebook for students and assignments using a 2D list.

**Requirements:**
- Build a 5×4 grid (5 students, 4 assignments) filled with `0`
- Write a function `set_score(gradebook, student, assignment, score)` that updates a score
- Write a function `print_gradebook(gradebook)` that prints each student’s scores neatly
- randomize grades selecting 4 in a range of 65-100  
- Use nested loops to compute and return tuple:
  - each student’s **average score**
  - the **overall class average**
  - Find the student (row) with the highest total score.
  - print returned tuple using an f-string




````python
# F3 answer here
import random

def set_score(gradebook, student, assignment, score):
  gradebook[student-1][score]=score

def print_gradebook(gradebook):
  for student in range(len(gradebook)):
    print(f"Student {student+1}: ", end="") #displays it nicely using an f string to tell the user which student it is
    for assignment in range(len(gradebook[0])):
      print(gradebook[student][assignment], end=" ")
    print()


grid = []
for row in range(5):              # outer loop = rows
    row = []
    for col in range(4):          # inner loop = columns
        row.append(0)
    grid.append(row)

for student in range(5):
  for assignment in range(4):
    grid[student][assignment] = random.randint(65, 100)

print_gradebook(grid)
student_avgs=[]
total_scores = 0
scores_count = 0
highest_student = 0

print()
for student in range(len(grid)):
  row_total = 0
  for score in grid[student]:
      row_total += score
      total_scores += score
      scores_count += 1
  student_avg = row_total / len(grid[student])
  student_avgs.append(student_avg)

highest_avg=student_avgs[0]
for index, score in enumerate(student_avgs):
  if score>highest_avg:
    highest_avg=score
    highest_student=index

overall_avg= total_scores/scores_count


results=(student_avgs, overall_avg, highest_student )


print(f"Student averages: {results[0]}")
print(f"Class average: {results[1]:.2f}")
print(f"Top student: Student {results[2]+1}")# added one to make sure that index matches up to actual student

````

- Use nested loops to count how many of your drawing symbols appear in the grid

**Optional extension:** Draw a border around the grid using `"#"`.
